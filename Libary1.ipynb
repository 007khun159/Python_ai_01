```python
#การ plot graph แบบคลื่น บีสต์
from matplotlib import rcParams

rcParams.update({'font.size':16})

x = np.arange(0, 10, 0.05 )
y1 = np.sin(5*x)
y2 = np.sin(6*x)
y = y1+y2
ybeat = 2*np.cos (0.5*(6-5)*x)


plt.xlabel('time (ms)')
plt.ylabel('signal(ms)')
plt.ylim(-2,4)

#สามารถเพิ่มสีได้ในตัวของ plot
plt.plot(x,y1,'gs',label = '1st wave')
#โชว์ตัวlabel
plt.legend()
#plt.scatter(x,y1,'green')
plt.show()
```

```python
#ทำการนำมาซ้อนกัน  

rcParams.update({'font.size':16})

x = np.arange(0, 10, 0.05)
y1 = np.sin(5*x)
y2 = np.sin(6*x)
y = y1 + y2
ybeat = 2*np.cos( 0.5*(6-5)*x )

plt.xlabel( 'time (ms)' )
plt.ylabel( 'signal' )
plt.ylim( -2, 4 )
plt.plot( x, y1, 'gs',label='1st wave' )
plt.legend()
#plt.show()

plt.xlabel( 'time (ms)' )
plt.ylabel( 'signal' )
plt.ylim( -2, 4 )
plt.plot( x, y2, 'ro',label='2nd wave' )
plt.legend()
plt.show()
#------------------------------------------#
plt.xlabel( 'time (ms)' )
plt.ylabel( 'signal' )
plt.ylim( -2, 4 )
plt.plot( x, y, 'k-',linewidth=5,label='1st+2nd wave' )
plt.plot( x, ybeat, 'b--',linewidth=2,label='beats' )
plt.legend()
plt.show()
#------------------------------------------#
```
